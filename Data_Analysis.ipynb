```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the text data from the CSV file into a pandas DataFrame
csv_filename = 'google_news.csv'

df = pd.read_csv(csv_filename)
print(df.shape)

# Remove any rows with missing values
df.dropna(inplace=True)
print(df.shape)

# Calculate paragraph lengths and add as a new column
df['Paragraph Length'] = df['Text'].apply(lambda x: len(x))

# Maximum length of a paragraph
max_length = df['Paragraph Length'].max()

# Minimum length of a paragraph
min_length = df['Paragraph Length'].min()

# Average length of paragraphs
avg_length = df['Paragraph Length'].mean()

# Median length of paragraphs
median_length = df['Paragraph Length'].median()

# Standard deviation of paragraph lengths
std_dev = df['Paragraph Length'].std()

# Display summary statistics
print(f"Maximum length of a paragraph: {max_length} characters")
print(f"Minimum length of a paragraph: {min_length} characters")
print(f"Average length of paragraphs: {avg_length:.2f} characters")
print(f"Median length of paragraphs: {median_length} characters")
print(f"Standard deviation of paragraph lengths: {std_dev:.2f} characters")

# Plot a histogram of paragraph lengths
df['Paragraph Length'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Paragraph Lengths')
plt.xlabel('Length of Paragraph')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## To analyse the text data

```python
from collections import Counter
# Concatenate all paragraphs into a single string
all_text = ' '.join(df['Text'])

# Split the text into words
words = all_text.split()

# Count the frequency of each word
word_freq = Counter(words)

# Get the most common words and their frequencies
most_common_words = word_freq.most_common(10)

# Display the most common words
print("Most common words:")
for word, freq in most_common_words:
    print(f"{word}: {freq}")

# Plot a bar chart of the most common words
plt.figure(figsize=(10, 6))
plt.bar(*zip(*most_common_words), color='skyblue')
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the text data from the CSV file into a pandas DataFrame
csv_filename = 'google_news.csv'

df = pd.read_csv(csv_filename)

# Remove any rows with missing values
df.dropna(inplace=True)

# Concatenate all paragraphs into a single string
all_text = ' '.join(df['Text'])

# Generate word cloud
wordcloud = WordCloud(width=1000, height=500, background_color='white').generate(all_text)

# Display the word cloud
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Wikipedia Text')
plt.axis('off')
plt.show()
```
